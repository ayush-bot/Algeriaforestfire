```python
import sqlite3
```

```python
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
Create Table if Not Exists employees(
               id integer Primary Key,
               name text not null,
               age integer,
               department text)
''')
connection.commit()
```

```python
cursor.execute('''
select * from employees
''')
```

```python
#insert the data in database 
cursor.execute('''
insert into employees(name,age,department)
               values('Ayush',24,'IITIAN')
''')

cursor.execute('''
insert into employees(name,age,department)
               values('deepa',24,'IITIAN')
''')

connection.commit()
```

```python
#querry the data from sqllite3
cursor.execute('''
select * from employees
''')
rows=cursor.fetchall()
for i in rows:
    print(i)
```
